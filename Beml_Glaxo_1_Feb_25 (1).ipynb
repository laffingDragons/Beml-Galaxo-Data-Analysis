```python
#importing libaries
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#Importing the 'BEML' dataset
beml_df = pd.read_csv('BEML.csv')
beml_df.head()
```

```python
glaxo_df = pd.read_csv('GLAXO.csv')
glaxo_df.head()
```

```python
#filtering the required columns
beml_df=beml_df[['Date','Close']]
glaxo_df=glaxo_df[['Date','Close']]
```

```python
#Getting the data types
beml_df.dtypes
```

```python
glaxo_df.dtypes
```

```python
#creating the index
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
beml_df.head()
```

```python
#creating the index
glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(beml_df['Date']))
glaxo_df.head()
```

```python
#Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize= (6,4))
plt.plot(glaxo_df.index, glaxo_df.Close)
plt.xlabel('Year')
plt.ylabel('Close Price')
plt.title('Closing Price of Glaxo')
plt.show()
```

```python
plt.figure(figsize= (6,4))
plt.plot(beml_df.index, beml_df.Close)
plt.xlabel('Year')
plt.ylabel('Close Price')
plt.title('Closing Price of Beml')
plt.show()
```

```python
#adding a 'gain' column
beml_df['gain'] = beml_df.Close.pct_change(periods = 1)
beml_df.head()
```

```python
glaxo_df['gain'] = glaxo_df.Close.pct_change(periods = 1)
glaxo_df.head()
```

```python
beml_df.dropna(inplace=True)
glaxo_df.dropna(inplace=True)
```

```python
beml_df.head()
```

```python
#plot gain
plt.figure(figsize= (9,4))
plt.plot(beml_df.index, beml_df.gain)
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of Beml')
plt.show()
```

```python
#plot gain for glaxo
plt.figure(figsize= (9,4))
plt.plot(glaxo_df.index, glaxo_df.gain)
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Gain of Glaxo')
plt.show()
```

```python
#2-feb-25
#Density plot for Glaxo
import seaborn as sns
sns.distplot(glaxo_df['gain'], label = "Glaxo", kde = True)
plt.xlabel('Gain')
plt.ylabel('Frequency')
plt.title('Desity plot for the Gaiin of Glaxo')
plt.show()
```

```python
sns.distplot(beml_df['gain'], label = "BEML", kde= True)
plt.xlabel('Gain')
plt.ylabel('Frequency')
plt.title('Desity plot for the Gaiin of BEML')
plt.show()
```

```python
#Getting the Statistic Proprtties
print("Mean : ", round(glaxo_df['gain'].mean(), 4))
print("Standard Deviation : ", round(glaxo_df['gain'].std(), 4))
```

```python
print("Mean : ", round(beml_df['gain'].mean(), 4))
print("Standard Deviation : ", round(beml_df['gain'].std(), 4))
```

```python
from termcolor import colored
from IPython.display import display, Markdown, HTML

# Define the statistics
glaxo_mean = round(glaxo_df['gain'].mean(), 4)
glaxo_std = round(glaxo_df['gain'].std(), 4)

beml_mean = round(beml_df['gain'].mean(), 4)
beml_std = round(beml_df['gain'].std(), 4)

# Generate the styled output
output = f"""
🌟 **Stock Statistics** 🌟
---
| 📈 **Stock**  | 🎯 **Mean Gain** | 📊 **Standard Deviation** |
|--------------|----------------|----------------|
| 🟢 **Glaxo** | `{glaxo_mean}`  | `{glaxo_std}`  |
| 🔵 **BEML**  | `{beml_mean}`   | `{beml_std}`   |

---
✅ *Mean Gain* represents the average return.
📉 *Standard Deviation* shows the risk (higher means more volatile).
"""

# Display as Markdown for rich formatting
display(Markdown(output))
```

```python
#getting the 95% confidence interval
from scipy import stats
glaxo_df_ci = stats.norm.interval(0.95, loc = glaxo_df.gain.mean(), scale = glaxo_df.gain.std())

#Confidence Interval
print('Glaxo Confidence Interval : ', np.round(glaxo_df_ci,4))
```
